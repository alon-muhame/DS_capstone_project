```python
import pandas as pd
import sklearn as sk
from pandas_profiling import ProfileReport
```

```python
unified_data = pd.read_csv("data/unified-data-per-country.csv")
unified_data.head()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
#dataset.drop('Confirmed cases', 1, inplace=True)
#dataset.drop('Deaths', 1, inplace=True)


corrMatrix = unified_data.corr(method='spearman')
print(corrMatrix)
plt.figure(figsize=(10, 8))
sn.heatmap(corrMatrix, annot=True, cmap=sn.diverging_palette(220, 10, as_cmap=True))
plt.show()
```

```python
corrMatrix = unified_data.corr(method='pearson')
plt.figure(figsize=(10, 8))
sn.heatmap(corrMatrix, annot=True, cmap=sn.diverging_palette(220, 10, as_cmap=True))
plt.show()
corrMatrix = unified_data.corr(method='spearman')
plt.figure(figsize=(10, 8))
sn.heatmap(corrMatrix, annot=True, cmap=sn.diverging_palette(220, 10, as_cmap=True))
plt.show()
```
